```python
#3
date = pd.date_range(start='7/13/2020', end='8/3/2020')
count_mask = [10, 15, 20, 18, 17, 14, 12, 13, 11, 9, 10, 8, 7, 5, 6, 4, 3, 2, 1, 0, 1, 2]
count_nomask = [25, 30, 28, 27, 26, 25, 24, 23, 22, 20, 19, 18, 17, 15, 14, 13, 12, 11, 10, 9, 8, 7]

DF = pd.DataFrame({
    'date': date,
    'count_mask': count_mask,
    'count_nomask': count_nomask
})

DF['date'] = pd.to_datetime(DF['date'])

plt.figure(figsize=(10, 6))
plt.plot(DF['date'], DF['count_mask'], label='Mask Counties', color='blue', marker='o')
plt.plot(DF['date'], DF['count_nomask'], label='No Mask Counties', color='red', marker='o')
plt.title('Kansas COVID-19 7-Day Rolling Average of Daily Cases (Per 100k Population)')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()


```
